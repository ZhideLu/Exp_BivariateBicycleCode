### Decoders

```python
script = "Decoder_18_4_4.py"

processes = []
for scaling_factor in list_scaling_factor:
    p = subprocess.Popen([sys.executable, script, str(scaling_factor)])
    processes.append(p)

# 等待所有进程完成
for p in processes:
    p.wait()
```
